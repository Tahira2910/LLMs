```python
from transformers import AutoTokenizer, AutoModelForCausalLM
 # Specify the desired download directory
model_path = "/content/sample_data/models"

tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-3.2-1B")
model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-3.2-1B")
```

```python
import torch
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
```

```python
def generate_text(prompt, max_new_tokens=100, temperature=0.7, top_p=0.9):
    input_ids= tokenizer(prompt, return_tensors="pt").input_ids.to("cuda")
    output_ids = model.generate(input_ids, max_new_tokens=max_new_tokens, temperature=temperature, top_p=top_p)
    output_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return output_text
```

```python
while True:
  user_input = input("User: ")
  if user_input.lower() == "exit":
    break
  response = generate_text(user_input)
  print("Llama 3.2: ", response)
```
